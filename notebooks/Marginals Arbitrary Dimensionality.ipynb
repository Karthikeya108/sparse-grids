```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import sys
sys.path.append('../')
import numpy as np
import scipy as sp
from scipy import stats
import matplotlib.pylab as pl
import seaborn as sns
import density_estimation.SGDensityEstimator as de
import pandas as pd
import pysgpp as ps
```

```python
dim = 10
size = 10000 * dim
mean = np.repeat(0.5, dim)
```

```python
# generate random covariance matrix
A = np.random.rand(dim, dim)/10
noise = A.T.dot(A)
cov = np.diag([0.0125]*dim) + noise
print cov
```

```python
# generate data 
rv = sp.stats.multivariate_normal(mean = mean, cov = cov)
data = rv.rvs(size)
```

```python
def filter_data(data):
    """Choose only entries between 0 and 1"""
    data_filtered = []
    for x in data:
        if np.min(x) >=0 and np.max(x) <= 1:
            data_filtered.append(x)
    return np.array(data_filtered)
```

```python
print "before filtering", data.shape
data = filter_data(data)
print "after filtering", data.shape
#data = (data-m)/(M-m)
```

```python
df = pd.DataFrame(data[:1000,:])
```

```python
pd.scatter_matrix(df, alpha=0.2, figsize=(10, 10), diagonal='kde')
```

```python
level = 3
regparam = 1e-4
regstr = 'laplace'
alpha_threshold = 0.5
simpling_size_first = 1000
sampling_size_last = 1000
epsilon = 1e-4

data_dm = ps.DataMatrix(data)

sgde = de.SG_DensityEstimator({'data':data_dm}, level, regparam, regstr, 
                           alpha_threshold=alpha_threshold,
                           sampling_size_first=simpling_size_first, 
                           sampling_size_last=sampling_size_last,
                           epsilon=epsilon, imax = 1)
    
grid, alpha = sgde.run()
```

```python
storage = grid.getStorage()
X = np.empty((grid.getSize(), dim), dtype=np.float32)
coords_dv = ps.DataVector(dim)
for i in xrange(grid.getSize()):
    idx = storage.get(i)
    idx.getCoords(coords_dv)
    X[i,:] = coords_dv.array()
X = X-mean
```

```python
covInvX = sp.linalg.solve(cov, X.T)
```

```python
nodal_values = - 0.5 * np.array([X[i,:].dot(covInvX[:,i]) for i in xrange(X.shape[0])])
opHierach = ps.createOperationHierarchisation(grid)
alpha = ps.DataVector(nodal_values)
opHierach.doHierarchisation(alpha)
sgde.alpha = alpha.array()
```

```python
sgde.sampling_size_first = 100000
```

```python
E_model = sgde.calc_model_expected_values(1)
```

```python
E_empir = sgde.calc_empirical_expected_values(data_dm)
```
